# Olasılık Hesabı standart sapma ile

### her sınıfın her özelliği için verileri tutan ve olasılık hesabı yapan obje sınıfı

```python
class features:
    def __init__(self, fname, fdata):
        self.fdata = [fdata]
        self.fname=fname
    def add_data(self,data):
        self.fdata.append(data)
    def stdrt_deviation(self):
        return stdev(self.fdata)
    def ret_mean(self):
        return statistics.mean(self.fdata)
    def probability(self,data):
        stsap=stdev(self.fdata)
        prob=(1/stsap*math.sqrt(2*math.pi)*math.exp((-1/2)*math.pow((data-statistics.mean(self.fdata))/stsap,2)))
        return prob
```

fclist => feature column list,
frlist => feature row list,

```python
fcList=[]
```

```python
for col in range(1,column):
    frList = []
    frList.append(features(ws.cell(2, column).value, ws.cell(2, col).value))
    for row in range(3, trainLen):
        IsIn = False
        for ftr in frList:
            if ftr.fname == ws.cell(row, column).value:
                ftr.add_data(ws.cell(row, col).value)
                IsIn = True
                break
        if IsIn != True:
            frList.append(features(ws.cell(row, column).value, ws.cell(row, col).value))
    fcList.append(frList)
```

### Her özellik ve her sınıf için ayrı ayrı standart sapma ve ortalama hesabı ile sıralayan

```python
print("Sınıf","Adet","Standart Sapma","Ortalama",sep='    ')
for fc in fcList:
    for fr in fc:
        print(fr.fname, len(fr.fdata), fr.stdrt_deviation(), fr.ret_mean(), sep=' ')
        print()
    print("****************************************************")
```

# Test Dataset Sınıf bulma ve Doğruluk Hesabı

```python
ttCount=0
```

```python
print("Beklenen Değer","Çıkan Sonuç",sep='    ')
for row in range(testStart,testLen):
    desList = []
    col = 1
    for fc in fcList:
        fname = ""
        data = ws.cell(row, col).value
        for fr in fc:
            if fname=="":
                fname = fr.fname
                prob = fr.probability(data)
            else:
                temp = fr.probability(data)
                if prob < temp:
                    prob = temp
                    fname = fr.fname
        desList.append(fname)
        col += 1
    cName=statistics.mode(desList)
    print(ws.cell(row,column).value,cName,sep=((18-len(ws.cell(row,column).value))*' '))
    if cName==ws.cell(row,column).value:
        ttCount+=1
```
