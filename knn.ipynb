### Satır indisini, uzaklığını ve sınıf adını tutan 

```python
class distance_class:
    def __init__(self,i,dist,categori_name):
        self.i=i
        self.dist=dist
        self.categori_name=categori_name
```

### k sayısına göre sınıfa karar verme etapında, k adet sınıfın aynı olanlarının sayısını tutan

```python
class categori_get:
    def __init__(self,categori_name,catCount):
        self.categori_name=categori_name
        self.catCount=catCount
    def inc_count(self):
        self.catCount=self.catCount+1
```

## Euclidean distance

```python
def distCalc(train_row, test_row, column):
    dist=0
    for c in range(1,column):
        dist+=pow((ws.cell(train_row, c).value)-(ws.cell(test_row, c).value),2)
    return math.sqrt(dist)
```

### k adet sınıftan en çok tekrar edeni gönderir

```python
def categori_decision(dList):
    cList=[]
    cList.append(categori_get(dList[0].categori_name,1))
    for i in range(1,len(dList)):
        IsIn=False
        for j in range(0,len(cList)):
            if dList[i].categori_name==cList[j].categori_name:
                cList[j].inc_count()
                IsIn=True
                break
        if IsIn!=True:
            cList.append(categori_get(dList[i].categori_name,1))
    cList.sort(key = lambda x : x.catCount)
    return cList[len(cList)-1].categori_name
```

## k=5 için hesaplama

```python
k=5
ttCount=0
```

```python
print("Beklenen Değer","Çıkan Sonuç",sep='    ')
for j in range(testStart,testLen):
    distList=[]
    for i in range(2,trainLen):
        #print(distCalc(i,j,column))
        distList.append(distance_class(i,distCalc(i,j,column),ws.cell(i,column).value))
    distList.sort(key = lambda x : x.dist)
    cName=str(categori_decision(distList[0:k+1]))
    print(ws.cell(j,column).value,cName,sep=(18-len(ws.cell(j,column).value))*' ')
    if cName==ws.cell(j,column).value:
        ttCount+=1
```
